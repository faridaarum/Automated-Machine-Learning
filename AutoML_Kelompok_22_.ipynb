# remove unneeded columns

```python
unneeded_columns =  ['id']
df_train = df_train.drop(columns=unneeded_columns)
```

```python
df_train.head()
```

# Handling Missing Value

```python
count_per_column = df_train.eq('?').sum()
print(count_per_column)
```

```python
df_train.replace("?",np.nan, inplace=True)
missingV_columns = df_train.columns[df_train.isna().any()]
count_of_MV = df_train[missingV_columns].isna().sum()
print(count_of_MV)
```

```python
count_of_MV = list(count_of_MV.keys())
```

```python
imputer = SimpleImputer(strategy = 'median', missing_values = np.nan)
imputer.fit(df_train[count_of_MV])
df_train[count_of_MV] = imputer.transform(df_train[count_of_MV])
```

pada missing handling value digunakan median sebagai nilai yang dipilih untuk mengisi kekosongan data. median sendiri dipilih karena jika dilihat pada data skewness :   

```python
skewness = df_train[missingV_columns].skew()
print(skewness)
```

Dengan demikian, semua nilai skewness pada data yang diberikan adalah positif, menunjukkan bahwa ekor distribusi data berada di sebelah kanan nilai pusat. menurut beberapasumber, makapendekatan yang baik adalah menggunakan median.

```python
df_train.info()
```

# Encoding

```python
diagnosis_mapping = {'M': 1, 'B': 0}
df_train['diagnosis'] = df_train['diagnosis'].map(diagnosis_mapping)
```

```python
df_train.head()
```

# Duplicated Data

```python
duplicated_data = df_train.duplicated()
duplicated_data_count = duplicated_data.value_counts()
print(duplicated_data_count)
```

# Removes columns that are not correlated with other features

```python
corr_matrix = df_train.corr()
```

```python
threshold = 0.5
low_correlation_columns = []

for column in corr_matrix.columns:
    correlations = corr_matrix[column].drop(column)
    if (correlations.abs() <= threshold).all():
        low_correlation_columns.append(column)

print(low_correlation_columns)
```

```python
print(corr_matrix)
```

# Split Data

```python
target = df_train["diagnosis"]
df_train.drop("diagnosis", axis=1, inplace=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_train, target, train_size=0.75, test_size=0.25)
```

```python
X_train.head()
```

```python
X_test.head()
```

# Modelling

```python
import time
import datetime
```

```python
import multiprocessing

start = time.time()
if __name__ == '__main__':
    multiprocessing.set_start_method('forkserver', force=True)
    tpot_model = TPOTClassifier(generations=10, 
                                population_size=50, 
                                verbosity=2, n_jobs=20, 
                                random_state=50, 
                                early_stop=3, 
                                mutation_rate=0.9, 
                                crossover_rate=0.1, 
                                subsample=0.8)
    tpot_model.fit(X_train, y_train)
end = time.time()
```

```python
# Convert the elapsed time to a timedelta object

from datetime import datetime

elapsed_time = datetime.fromtimestamp(end) - datetime.fromtimestamp(start)
print("Elapsed Time: ", elapsed_time)
```

```python
print(tpot_model.score(X_test, y_test))
```
